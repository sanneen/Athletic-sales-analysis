```python
# Import Libraries and Dependencies
import pandas as pd
```

```python
# Read the CSV files into DataFrames.
a2020 = pd.read_csv("athletic_sales_2020.csv")
a2021 = pd.read_csv("athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
a2020
```

```python
# Display the 2021 sales DataFrame
a2021
```

```python
# Check the 2020 sales data types.
a2020.info()
```

```python
# Check the 2021 sales data types.
a2021.info()
```

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic = pd.concat([a2020, a2021], axis="rows", join="inner")
athletic.reset_index(drop=True, inplace=True)
athletic
```

```python
# Check if any values are null.
athletic.isnull().sum()
```

```python
# Check the data type of each column
athletic.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic['invoice_date'] = pd.to_datetime(athletic["invoice_date"])
athletic
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic.info()
```

```python
# Determine which Region Sold the Most Products
# Using groupby

# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# Show the top 5 results.

totalproduct= athletic.groupby(["region","state","city"])[["units_sold"]].sum().sort_values(by="units_sold", ascending=False)
totalproductrename=totalproduct.rename(columns={"units_sold":"Total_Products_Sold"})
totalproductrename.head(5)
```

```python
# Using pivot_table
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold"
# Show the top 5 results.

tppivot = pd.pivot_table(athletic,
                       index=["region","state","city"],
                       values="units_sold",
                       aggfunc='sum').sort_values(by="units_sold", ascending=False)
tprename=tppivot.rename(columns={"units_sold":"Total_Products_Sold"})
tprename.head(5)
    
```

```python
# Determine which Region had the Most Sales
# Using groupby
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

totalsales= athletic.groupby(["region","state","city"])[["total_sales"]].sum().sort_values(by="total_sales", ascending=False)
totalsalesrename=totalsales.rename(columns={"total_sales":"Total Sales"})
totalsalesrename.head(5)

```

```python
# Using pivot_table
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

tspivot = pd.pivot_table(athletic,
                       index=["region","state","city"],
                       values="total_sales",
                       aggfunc='sum').sort_values(by="total_sales", ascending=False)
tsrename=tspivot.rename(columns={"total_sales":"Total Sales"})

tsrename.head(5)
    
```

```python
# Determine which Retailer had the Most Sales
# Using groupby
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

retailgroupby= athletic.groupby(["retailer","region","state","city"])[["total_sales"]].sum().sort_values(by="total_sales", ascending=False)
retailrename=retailgroupby.rename(columns={"total_sales":"Total Sales"})
retailrename.head(5)
```

```python
# Using pivot_table
# Show the total sales for the products sold for each retailer, region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results

rpivot = pd.pivot_table(athletic,
                       index=["retailer","region","state","city"],
                       values="total_sales",
                       aggfunc='sum').sort_values(by="total_sales", ascending=False)
rrename=rpivot.rename(columns={"total_sales":"Total Sales"})

rrename.head(5)
```

```python
# Determine which Retailer Sold the Most Women's Athletic Footwear
# Filter the sales data to get the women's athletic footwear sales data.
women=athletic.loc[athletic["product"]=="Women's Athletic Footwear"]
women
```

```python
# Using groupby
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.

wgroupby= women.groupby(["retailer","region","state","city"])[["units_sold"]].sum().sort_values(by="units_sold", ascending=False)
wrename=wgroupby.rename(columns={"units_sold":"Womens Footwear Units Sold"})
wrename.head(5)
```

```python
# Using pivot_table
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.

wpivot = pd.pivot_table(women,
                       index=["retailer","region","state","city"],
                       values="units_sold",
                       aggfunc='sum').sort_values(by="units_sold", ascending=False)
wpivotrename=wpivot.rename(columns={"units_sold":"Womens Footwear Units Sold"})

wpivotrename.head(5)

```

```python
# Determine the Day with the Most Women's Athletic Footwear Sales
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.

womenpivot = pd.pivot_table(women,
                       index=["invoice_date"],
                       values="total_sales",
                       aggfunc='sum').sort_values(by="total_sales", ascending=False)
womenrename=womenpivot.rename(columns={"total_sales":"Total Sales"})
womenrename.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".

womendaily=womenrename.resample("D").sum().sort_values(by = "Total Sales",ascending = False)
womendaily.head(10)
```

```python
# Determine the Week with the Most Women's Athletic Footwear Sales
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".

womenweekly=womenrename.resample("W").sum().sort_values(by = "Total Sales",ascending = False)
womenweekly.head(10)
```
